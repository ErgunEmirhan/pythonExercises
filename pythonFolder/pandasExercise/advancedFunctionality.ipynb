```python
import pandas as pd
```

```python
bios = pd.read_csv('bios.csv')
bios
```

```python
coffee = pd.read_csv('coffee.csv')
coffee['yesterday_revenue'] = coffee['Units Sold'].shift(2)
coffee
```

```python
coffee['pct_change'] = coffee['Units Sold'] / coffee['yesterday_revenue'] * 100
coffee
```

```python
bios['height_rank'] = bios['height_cm'].rank() # .rank(ascending=False) will make the first person rank 1 I guess
bios.sort_values(['height_rank'], ascending=True)
```

```python
# Cumulative sum

coffee['Units Sold'].cumsum().reset_index()
```

```python
latte = coffee[coffee['Coffee Type'] == 'Latte'].copy()
latte['3day'] = latte['Units Sold'].rolling(3).sum() 
# rolling3 creates a window of length 3 which means it will calculate the sum of the current value and the previous two values
latte
```
