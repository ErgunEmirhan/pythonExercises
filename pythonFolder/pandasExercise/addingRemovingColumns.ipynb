# Adding/Removing Columns

```python
import pandas as pd
import numpy as np
```

```python
coffee = pd.read_csv('coffee.csv')
```

```python
coffee_new = coffee.copy() 
# if coffee_new = coffee, then coffee_new is a reference to coffee so the below line changes the values of coffee dataframe also
# here we used coffee.copy() to create a new dataframe so original dataframe is not modified
coffee_new['price'] = 4.99
coffee
```

```python
coffee['new_price'] = np.where(coffee['Coffee Type'] == 'Espresso', 3.99, 5.99)
```

```python
coffee.drop(columns=['price']) #  returns the modified dataframe but the original dataframe is not modified
```

```python
bios = pd.read_csv("bios.csv")
bios_new = bios.copy()
```

```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'short' if x < 165 else ('Average' if x < 185 else 'Tall'))
# with apply, we can apply a function to each row of the dataframe
bios.head()
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 and row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'
    
bios['weight_category'] = bios.apply(categorize_athlete, axis=1)
bios.head()
```
