# Loading in Dataframes from Files

```python
import pandas as pd
```

```python
coffee = pd.read_csv('./coffee.csv')  # pd.read_parquet() for parquet file, pd.read_excel(path, sheet_name='sheet1') for Excel
```

```python
coffee.head()
```

```python
coffee.sample(3) # give 3 random sample, coffee.sample(3, random_state=1) for fixed sample
```

```python
# loc and iloc
# coffee.loc[#rows,#columns]
coffee.loc[0] # just first day
coffee.loc[[0, 1, 2]] # first 3 days
coffee.loc[1:6, ['Day', 'Units Sold']] # day and units sold
# if coffee.index = coffee["Day"], then you should use coffee.loc["Monday":"Thursday"] etc.
```

```python
coffee.iloc[1:6, [0, 2]] # day and units sold
# uses always index values
```

```python
coffee.loc[1:3, "Units Sold"] = 10 # update values
```

```python
coffee.at[0, "Units Sold"] = 10 # update specific value
coffee.iat[0, 2] = 10 # update specific value
```

```python
coffee.Day # since 'Units Sold' has space, we cannot call it like this
```

```python
coffee.sort_values(["Units Sold", "Coffee Type"], ascending=[False, True]) # sort in asc/desc orders
```

```python
# iterating but bad performance, not good practice generally, not optimized/ optimal
for index, row in coffee.iterrows():
    print(index)
    print(row) # print(row["Units Sold"])
    print("\n\n\n\n\n")
```
