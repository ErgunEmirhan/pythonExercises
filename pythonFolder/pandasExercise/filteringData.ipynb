# Filtering Data

```python
import pandas as pd
```

```python
bios = pd.read_csv('bios.csv')  
```

```python
bios.loc[bios['height_cm'] > 200, ['name', 'height_cm']] # with bios['height_cm] > 200, we filtered the results
bios[bios['height_cm'] > 200][['name', 'height_cm']] # filtering again

# what about multiple filtering?
bios[(bios['height_cm'] > 215) & (bios['born_country'] == 'USA')][['name', 'height_cm']]

bios[bios['name'].str.contains("Keith")][['name', 'height_cm']] # filtering by string contains
bios[bios['name'].str.contains("keith", case=False)][['name', 'height_cm']] # filtering by string contains, case insensitive
bios[bios['name'].str.contains("keith|patrick", case=False)][['name', 'height_cm']] # filtering by string contains, case insensitive, logical or used
# you can also use regex
bios[bios['born_date'].str.contains(r'^19', na=False)][['name', 'height_cm']] # find athletes born in a year starting with 19
bios[bios['name'].str.contains("keith|patrick", case=False, regex=False)][['name', 'height_cm']] 
# filtering by string contains, case insensitive, since regex is false; it will search for the string 
bios[bios['born_country'].isin(['USA', "FRA", "GBR"])] # filtering by multiple values
```

```python
bios.query('born_country == "USA" and born_city == "Seattle"') # filtering by query NOTICE "" INSIDE '' 
```
